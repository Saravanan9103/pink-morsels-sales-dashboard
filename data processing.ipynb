```python
## Import Libraries
import pandas as pd
import numpy as np
from pandas import to_datetime
```

```python
## Read three csv files & combine as DF
sales_2018=pd.read_csv('../data/daily_sales_data_0.csv')
sales_2019=pd.read_csv('../data/daily_sales_data_1.csv')
sales_2020=pd.read_csv('../data/daily_sales_data_2.csv')
data = pd.concat([sales_2018, sales_2018,sales_2018], ignore_index=True)
```

```python
print(data.head())
```

```python
## Exploring Data
print(data.info())
```

```python
## Filter Pink Morsels
filtered_data=data[data['product']=='pink morsel']
print(filtered_data.head(20))
```

```python
## Change price column datatype to float
filtered_data["price"]=filtered_data["price"].str.replace("$", "", regex=False).astype(float)
```

```python
## Create Sales Column & convert date Column to datetime
filtered_data["sales"]=filtered_data["price"]*filtered_data["quantity"]
filtered_data["date"]=pd.to_datetime(filtered_data["date"])
```

```python
# Keep only Sales, Date, and Region
df=filtered_data[['sales', 'date','region']]
print(df.head(10))
```

```python
## 7. Export Processed Data
df.to_csv('formatted_sales_data.csv')
```
